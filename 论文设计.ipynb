**1. 系统概览与输入表征 (System Overview & Input Representation)**

**本系统旨在解决机械臂在非结构化流体环境中，从水下抓取物体并跨越“水-气”界面的高动态控制难题。核心挑战在于界面段（Interface Phase）的非平稳扰动与传感器物理特性的剧烈漂移。为此，我们提出了一种\*\*因果解耦（Causally Decoupled）\*\*的层次化架构，将物理环境感知与物体内容表征在信号源头进行分离，确保策略网络在极端的跨域条件下仍能获得稳健的决策依据。**

**1.1 时间窗口协议 (Window-based Protocol)**

**由于视觉与触觉来源于不同采样机制，二者在时间轴上天然存在异步性。本文中，视觉数据以视频形式采集，帧率固定为 30 Hz；触觉数据以文本流形式记录，不提供显式时间戳。经实测，触觉采样周期约为 39 ms，对应实际采样频率约为 25.64 Hz。若直接进行逐帧级配准，不仅会受到多速率采样差异的影响，也会在跨越水-气界面时因局部瞬态扰动而引入额外噪声。因此，本文放弃单帧严格对齐，采用\*\*统一时间窗口（Sliding Window）\*\*作为多模态处理的基本单元。**

**为建立统一时基，系统以夹爪开始闭合的时刻作为视觉与触觉的共享时间原点，记为 t=0。对于视频流，第 i 帧对应的时间戳记为**



**t\_i^{vid}=\frac{i}{30}**

**对于触觉流，由于原始文本记录不提供显式时间戳，本文依据固定采样周期构造伪时间戳序列。设第 j 行触觉数据对应的时间为**



**t\_j^{tac}=j\times 0.039**

**若索引从 1 开始，则等价写为 (j−1)×0.039。在当前实现中，不再额外引入固定偏移修正项，而依赖窗口级时序建模吸收小尺度跨模态时延误差。**

**在此统一时基下，定义长度为 LLL 的滑动窗口**



**W\_k=[t\_k^{s},\,t\_k^{e}), \quad t\_k^{e}=t\_k^{s}+L**

**本文最终采用固定窗口参数：**

**L=1.0s,S=0.25s,**

**其中 S 为窗口步长。第 k 个窗口的起始时间满足**



**t\_k^{s}=k⋅0.25.**

**仅保留满足的完整窗口，其中 t\_{valid\\_end} 表示当前样本在视频与触觉两种模态上均有效的共同结束时刻。**

**对于每个窗口 W\_k​，系统分别收集落入该时间区间内的全部视觉帧与全部触觉采样点，构成：**



**VWk={Ii∣tivid∈[tks,tke)},V\_{W\_k}=\{I\_i\mid t\_i^{vid}\in [t\_k^{s},t\_k^{e})\},VWk​​={Ii​∣tivid​∈[tks​,tke​)}, TWk={xj∣tjtac∈[tks,tke)}.T\_{W\_k}=\{x\_j\mid t\_j^{tac}\in [t\_k^{s},t\_k^{e})\}.TWk​​={xj​∣tjtac​∈[tks​,tke​)}.**

**在 1.0 s 的窗口长度下，一个窗口通常包含约 30 帧视频与 25–26 个触觉采样点。二者数量不同属于正常现象，本文不对两模态做强制重采样，而是在各自模态内部完成时序编码与窗口级聚合，最终在窗口表征层实现跨模态对齐与融合。**

**采用窗口协议而非逐帧配准具有三点优势。首先，它显式规避了 30 Hz 视频与约 25.64 Hz 触觉之间的不完全同步问题；其次，它能够在局部时间范围内保留界面过渡阶段的动态趋势，使模型更容易感知浮力变化、接触重分配与视觉扰动；最后，它为后续的三相标签生成、稳定段筛选以及跨介质对齐提供了统一且可操作的时间组织基础。对于本文关注的跨水-气界面抓取任务而言，窗口长度 1.0 s 足以覆盖局部接触动态，而 0.25 s 的步长又能够在约 2.5 s 的界面持续过程中产生足够密集的过渡窗口，从而兼顾表征稳定性与阶段分辨率。**

**1.2 频域信号预处理与物理分离 (Frequency Decomposition)**

**为了从源头上解决“介质判别”与“物体识别”的特征冲突，我们在进入神经网络之前，对原始触觉信号 $T\_W$ 实施显式的频域分离。**

**AC 分量（动态流）：**


**$$T\_{high} = T\_W - \text{EMA}(T\_W, \alpha)$$**


**物理含义： 通过高通滤波或差分提取。该分量去除了静水压力等低频偏置，**保留接触状态的动态变化分量（例如微滑移先兆、法向/切向载荷的短时波动、接触建立/破水瞬间的瞬态变化），用于刻画接触语义与稳定性风险。


**流向： 仅输入至 Content Encoder，用于提取物体语义与接触状态。**


**DC 分量（证据流）：**


**$$T\_{low} = \text{EMA}(T\_W, \alpha)$$**


**物理含义： 通过指数移动平均（EMA）或低通滤波提取。该分量显式保留了准静态载荷（Quasi-static Load）、静水压力偏置及其在出水过程中的卸载趋势（Unloading Trend）。**


**流向： 仅输入至 Evidence Encoder，作为介质信念推断的物理证据。**

**1.3 总体逻辑拓扑：三流架构 (Three-Stream Topology)**

在完成统一时间窗口构造后，系统围绕每个窗口 W 建立三条逻辑上解耦、功能上互补的处理路径：语义流（Semantic Stream）、物理流（Physics Stream）与任务流（Task Stream）。其中，任务流负责将对象的控制相关属性与介质状态联合引入策略层，从而生成面向跨介质抓取过程的期望力输出。

与传统做法中直接预测单一对象类别不同，本文将对象表示为一组**控制导向属性（control-oriented attributes）**。这样做的原因在于：对于跨水-气界面的抓取控制而言，真正影响策略调节的并非物体语义类别本身，而是其对接触建立、负载变化、滑移风险与安全力窗口具有直接作用的物理和几何属性。基于这一考虑，本文将对象属性分解为三个互补的属性轴：

**Fragility / Compliance**：fragile / robust / compliant

**Geometry**：cylindrical / bowl-like / cup-like / constricted-opening

**Surface**：smooth / matte

其中，Fragility / Compliance 描述对象在施力过程中的易碎性与柔顺性；Geometry 描述其主导几何形态及开口约束方式；Surface 则用于表征表面摩擦和接触纹理特征。这种多属性表示贴近抓取控制的真实需求。例如，一个迷你玻璃花瓶可被描述为 “fragile + constricted-opening + smooth”，而一个软质硅胶圆柱可被描述为 “compliant + cylindrical + smooth”。由此，策略网络响应的是对象的**控制条件组合**，而非对特定实例身份的机械记忆。

在实现上，任务流接收来自语义流的内容特征z\_{content}​（以及可选视觉上下文特征 z\_W^v​），通过多头属性预测器分别输出三个属性轴上的概率分布：



qW(f)=hf(zcontent,zWv),q\_W^{(f)} = h\_f(z\_{content}, z\_W^v),qW(f)​=hf​(zcontent​,zWv​), qW(g)=hg(zcontent,zWv),q\_W^{(g)} = h\_g(z\_{content}, z\_W^v),qW(g)​=hg​(zcontent​,zWv​), qW(s)=hs(zcontent,zWv).q\_W^{(s)} = h\_s(z\_{content}, z\_W^v).qW(s)​=hs​(zcontent​,zWv​).

为反映属性判断的可靠程度，系统进一步计算每个属性轴预测分布的熵作为不确定性度量：

H\_f(W), H\_g(W), H\_s(W).

最终，任务流将三组属性分布及其不确定性拼接并投影为对象控制属性特征



gW=Proj([qW(f),qW(g),qW(s),Hf(W),Hg(W),Hs(W)]).g\_W=\text{Proj}\big([q\_W^{(f)}, q\_W^{(g)}, q\_W^{(s)}, H\_f(W), H\_g(W), H\_s(W)]\big).gW​=Proj([qW(f)​,qW(g)​,qW(s)​,Hf​(W),Hg​(W),Hs​(W)]).

在策略阶段，任务流并不单独决定控制输出，而是与物理流给出的介质信念 p(W) 共同作用。前者回答“当前抓取对象在控制上具有什么属性”，后者回答“当前系统正处于水下、界面还是空气中”。二者联合决定策略应如何调节施力。具体而言，任务流输出的 g\_W​ 用于提供对象级控制先验，例如脆性器皿需要更保守的力增益、收口容器在界面附近可能伴随更复杂的残水与姿态变化、磨砂表面相较光滑表面具有不同的摩擦与接触响应；与此同时，物理流输出的介质信念 p(W) 通过条件化机制对策略网络进行阶段性调制，使系统在接近界面与完全出水后能够生成不同的期望力调整趋势。

因此，任务流的本质并不是“对象分类”，而是建立从**内容表征 → 控制属性组合 → 条件化策略调节**的映射关系。它使得系统能够将对象本体信息与跨介质状态变化解耦开来：对象“是什么样的”由g\_W表示，环境“当前处于哪一物理阶段”由 p(W) 表示，二者共同驱动最终的期望力输出

F^{des}(W).

这一设计使策略学习不再依赖对具体对象实例的死记硬背，而是基于对象控制属性与介质变化的因果组合关系进行响应，从而提升对未见器皿/容器实例的泛化能力。

**2. 解耦感知支路设计 (Decoupled Perception Streams)**

**本章节定义了三条并行特征提取路径的具体网络结构。所有编码器均以时间窗口 $W$ 为处理单元，通过时序注意力机制（Temporal Attention）将序列数据聚合为紧凑的窗口级表征。**

**2.1 视觉上下文编码 (Visual Context Encoding)**

**视觉模块通过两级映射结构，实现从原始像素到跨模态共享空间的表征转换。**

**骨干网络 (Backbone): 采用预训练的 ViT-B/16 (CLIP Image Encoder)。在完整设计中，视觉分支通过 LoRA 适配水下光学环境（如折射、浑浊及反光）；在当前工程实现中，Stage 1 保持视觉 backbone 冻结且不注入 LoRA，Stage 2 与 Stage 4 开启视觉 LoRA，用于视触对齐与联合微调，Stage 3 保留 LoRA 结构，但与 visual encoder 一起冻结。**

**特征提取: 提取窗口内每一帧的 [CLS] token。经过时序注意力池化（Temporal Attention Pooling）后，得到视觉窗口特征 $h\_W^v \in \mathbb{R}^{d\_{vit}}$。**

**投影对齐层 (Projection Head): 为了将视觉特征映射到视触共享语义空间，同时避免对比损失对主干网络特征空间的过度压缩，我们引入一个可训练的投影头：**

**$$z\_{W\\_raw}^v = \text{MLP}\_{proj}(h\_W^v)$$**

**该投影头由两层线性映射及非线性激活函数（如 ReLU 或 GELU）组成。**

**单位化处理 (Normalization): 在进入对比损失计算前，对投影后的向量进行 $L\_2$ 归一化，确保特征分布在单位超球面上，**得到最终用于共享语义空间对齐的视觉表征**：**



**$$z\_W^v = \frac{z\_{W\\_raw}^v}{\|z\_{W\\_raw}^v\|\_2}$$**

**特征使用约定：**
本文统一约定：

z\_W^v用于跨模态对齐与共享语义空间中的表征学习；

h\_W^v 作为保留更完整感知信息的窗口级视觉上下文特征，输入属性推断模块、任务上下文聚合模块与策略网络。

**2.2 触觉内容编码器 (Tactile Content Encoder - AC Branch)**

**该支路负责从高频动态分量 $T\_{high}$ 中提取精细的接触语义（如微滑移、纹理振动），并作为后续策略输入的关键感知依据。**

**输入: 窗口内的高频触觉序列 $T\_{high} \in \mathbb{R}^{L \times N \times 3}$，其中 $L$ 为窗口长度，$N$ 为传感器节点数。**


**网络结构 (PIF-style Spatio-Temporal Network):**


**空间投影 (Spatial Projection): 使用共享权重的 MLP 或 Graph Layer 对每一时刻的 $N$ 点阵列进行空间特征提取，捕获接触面的局部形变模式。**


**时序编码 (Temporal Encoding): 采用多层 Transformer Encoder 或 Causal Conv1D，捕捉接触状态随时间的动态演变（Dynamics）。**


**关键约束 (Causal Independence):**

**该编码器严禁接收任何形式的介质信念 $p(W)$ 或低频环境偏置 $T\_{low}$ 作为输入。**


**设计意图： 确保生成的表征 $z\_{content}$ 仅由物体交互本身的物理属性决定。这使得在后续的消融实验中，替换信念变量不会导致感知表征发生底层漂移（Drift），从而保证了因果验证的纯净度。**


**输出表征 ($z\_{content}$): 经时序聚合后，生成内容向量 $z\_{content} \in \mathbb{R}^{d}$，该向量将在后续与 $z\_W^v$ 进行对比学习对齐。**

**2.3 触觉证据编码器 (Tactile Evidence Encoder - DC Branch)**

**该支路旨在从低频分量 $T\_{low}$ 中“蒸馏”出物理环境的基线特征，专用于介质信念的推断。**

**输入: 窗口内的低频/准静态触觉序列 $T\_{low}$。**


**网络结构 (Lightweight Encoder):**

**考虑到低频信号的信息密度较低且变化缓慢，采用轻量级的 1D Convolutional Network 或简单的 MLP Mixer。**


**重点捕获信号的均值漂移 (Mean Shift) 和 长期趋势 (Long-term Trend)（例如出水过程中静水压力的线性卸载）。**


**功能边界:**

**该特征 $z\_{med}$ 不参与 跨模态对齐，也不直接输入策略网络。它仅作为后续 GRU 模块的物理证据输入。**


**输出表征 ($z\_{med}$): 聚合后的环境证据向量 $z\_{med} \in \mathbb{R}^{d\_{med}}$。**

**2.4 窗口级时序聚合 (Window-level Temporal Aggregation)**

**为了应对界面段（Interface Phase）的非平稳性，简单的平均池化（Mean Pooling）可能会平滑掉关键的瞬态线索（如破水瞬间的冲击）。系统对上述所有支路统一采用 时序注意力池化 (Temporal Attention Pooling)。**

**机制:**

**对于窗口内的特征序列 $H = \{h\_1, ..., h\_L\}$，计算加权和：**


**$$z\_W = \sum\_{t=1}^{L} \alpha\_t h\_t, \quad \text{where } \alpha\_t = \text{Softmax}\left(\frac{h\_t Q^T}{\sqrt{d\_k}}\right)$$**

**其中，$Q \in \mathbb{R}^{1 \times d}$ 是一个可学习的全局查询向量 (Learnable Global Query)。与传统的平均池化不同，$Q$ 通过反向传播学习任务相关的特征模态，从而在窗口序列内动态分配权重。这使得系统能够自适应地聚焦于具有显著物理语义的时刻（如触觉载荷突变点），同时抑制界面段随机扰动产生的特征噪声。**

**作用: 网络能够自动关注窗口内“信息量最大”的时刻（例如接触发生瞬间或滑移起始瞬间），生成语义更稳定的窗口级 Embedding。**

**3. 跨模态对齐与共享语义学习 (Cross-Modal Representation Learning)**

**为了克服水下视觉的域偏移与触觉的介质噪声，本模块在隐空间中构建了一个分层映射架构。其核心思想是将特征空间划分为主干表征空间 (Backbone Space) 与 对齐投影空间 (Alignment Space)，从而在实现跨模态语义关联的同时，保护感知特征的物理完整性。**

**3.1 对称投影头设计 (Symmetric Projection Heads)**

**在视觉与触觉 AC 支路的编码器后端，分别引入结构对称的非线性投影头，将感知特征映射至单位超球面上：**

**投影映射：**


**$$e\_v = \text{L2Norm}(\text{MLP}\_v(z\_W^v)), \quad e\_t = \text{L2Norm}(\text{MLP}\_t(z\_{content}))$$**


**设计意图： 投影头（Projection Head）负责学习跨模态的坐标转换，而主干特征 $z\_{content}$ 则专注于保留原始的接触语义。这种解耦设计能防止对比损失对骨干网络特征空间的过度压缩（Collapse）。**

**3.2 视触模态对齐 (Visual-Tactile Alignment - $\mathcal{L}\_{clip}$)**

**利用 CLIP 范式的对比学习，在投影空间 (Alignment Space) 中建立视觉几何与触觉动态线索的关联。**

**目标： 确保当触觉感受到特定物理属性（如边缘曲率）时，其投影向量 $e\_t$ 与视觉观察到相应结构时的投影向量 $e\_v$ 在高维超球面上具有最大的余弦相似度。**


**损失函数： 采用 InfoNCE Loss。对于 Batch 中的 $N$ 个样本，计算投影向量间的对齐损失：**


**$$\mathcal{L}\_{clip} = -\sum\_{i=1}^{N} \log \frac{\exp(\text{sim}(e\_t^{(i)}, e\_v^{(i)}) / \tau)}{\sum\_{j=1}^{N} \exp(\text{sim}(e\_t^{(i)}, e\_v^{(j)}) / \tau)}$$**

**3.3 跨介质鲁棒性正则 (Cross-Medium Robustness - $\mathcal{L}\_{inv}$)**

**为使感知表征对环境扰动具备内在鲁棒性，我们在主干表征空间 (Backbone Space) 直接施加基于对象身份 (Level-2 Object ID) 的监督对比学习 (SupCon)。**

**正样本对定义： 选取同一对象 ID 在\*\*“稳定水下段”与“稳定空气段”\*\*的窗口样本互为正对。**


**物理直觉： 在特征生成的源头强制执行介质不变性约束，使编码器学会自动滤除流体阻尼等环境噪声，聚焦于对象本质的物理属性（如刚度、摩擦特性）。**


**损失函数：**


**$$\mathcal{L}\_{inv} = \sum\_{i \in I} \frac{-1}{|P(i)|} \sum\_{p \in P(i)} \log \frac{\exp(z\_{content}^{(i)} \cdot z\_{content}^{(p)} / \tau)}{\sum\_{a \in A(i)} \exp(z\_{content}^{(i)} \cdot z\_{content}^{(a)} / \tau)}$$**


**注：由于 $z\_{content}$ 将直接输入策略网络，该损失确保了下游决策获得的是预先“去噪”的纯净语义。**

**3.4 稳定性感知采样策略 (Stability-Aware Sampling Strategy)**

**为防止非平稳物理信号污染表征空间，系统实施以下采样协议：**

**界面段屏蔽 (Interface Masking)： 在计算 $\mathcal{L}\_{inv}$ 时，显式屏蔽标记为“界面段”的窗口。界面段的复杂波动（如破水冲击）仅用于介质信念 $p(W)$ 的推断，不参与不变性特征的对齐。**


**结构化 Batch 构造： 每次迭代随机抽取 $N\_{obj}$ 个对象 ID，并确保每个对象在水下和空气稳定段均有 $M$ 个窗口采样。这保证了 $\mathcal{L}\_{inv}$ 始终拥有跨介质的正样本对，防止训练梯度消失。**

**3.5 因果解释性保证 (Causal Interpretation Guarantee)**

**本模块的设计为论文的因果论证提供了数学支撑：**

**感知独立性： $z\_{content}$ 的生成完全独立于介质信念 $p(W)$。**


**表征稳定性： 通过 $\mathcal{L}\_{inv}$，系统在感知层实现了对介质扰动的鲁棒性，而非依赖策略层的后期补偿。**


**消融逻辑： 在测试中人为替换或冻结 $p(W)$ 时，$z\_{content}$ 保持不变。因此，任何抓取性能的下降均可排他性地归因于策略网络对物理逻辑切换（如重力补偿失效）的判定失败，而非由于“看不清”或“摸不准”导致的感知溃散。**

**4. 层次化状态与属性推断 (Hierarchical State & Attribute Inference)**

**本模块承担着“承上启下”的关键作用：它将底层的感知 Embedding 转化为策略网络可直接理解的高级状态变量。**设计核心在于通过递归状态估计与结构化条件推断，将底层感知 embedding 转化为策略可用的高级状态变量。

**4.1 递归介质信念推断 (Recurrent Medium Belief Inference)**

跨水-气界面抓取的核心挑战之一，在于系统需要在连续执行过程中实时判断当前窗口所处的物理介质状态。与依赖视觉水面检测或显式几何定位的方法不同，本文将介质判断完全建立在低频触觉证据及其时间演化之上，构建一个基于递归时序建模的介质信念推断模块。其目标并非对单帧观测进行瞬时分类，而是在连续时间窗口上估计系统处于 Water、Interface 和 Air 三种介质状态的概率分布，从而为后续策略提供连续、可微且可递归更新的状态条件。

设按时间排序的窗口序列为 {W\_t\}，其中当前窗口W\_t的低频触觉输入记为 T\_{low,t}。首先，经由触觉证据编码器对其进行编码，得到低频介质证据特征

z\_{med,t}=E\_{med}(T\_{low,t}).

该特征主要保留缓慢变化的接触偏置、负载迁移趋势、浮力衰减或消失过程中的整体力水平变化，以及出水前后由介质切换引发的低频动力学漂移。与高频触觉内容表征不同，z\_{med,t}不强调细粒度纹理或瞬态微振动，而聚焦于足以表征介质变化的全局物理趋势。为保持因果分工的清晰性，该模块仅接收低频触觉证据作为输入，不引入高频内容特征或对象属性信息。

在时序建模上，系统采用 GRU 递归整合历史窗口的低频证据。记上一时刻隐藏状态为h\_{t-1}，则当前窗口的递归状态更新为

h\_t=\mathrm{GRU}(z\_{med,t},\,h\_{t-1}),

并进一步通过一个轻量分类头得到当前窗口的介质 logits：

\ell(W\_t)=\mathrm{MLP}\_{med}(h\_t).

最终，经 Softmax 可得到窗口W\_t的介质信念分布：

p(W\_t)=\mathrm{Softmax}(\ell(W\_t)),

其中

p(W\_t)=\left[p\_{water}(W\_t),\,p\_{interface}(W\_t),\,p\_{air}(W\_t)\right].

这种递归结构使模型能够在局部噪声、液面扰动及瞬时接触波动存在时，仍然依据多窗口上下文形成稳定判断，而不是被单次触觉突变误导。尤其在界面过渡阶段，浮力衰减、附水残留、局部姿态变化与接触重分配往往是一个持续过程，递归记忆有助于模型从“趋势”而非“瞬时点”上识别介质变化。由于该推断过程不依赖显式几何位置、水面标定或额外位姿记录，它更适用于仅具备视觉与触觉观测的跨介质抓取设定。

与硬切换式的介质标签不同，p(W\_t)是一个连续概率分布，这一点对界面阶段尤为重要。因为从物体开始预备出水到完全脱离水体之间，本质上是一个持续的物理过渡过程，使用概率分布比单一离散状态更能刻画这种连续性。后续策略模块正是基于该介质信念，而非某个刚性阈值，对期望力输出进行连续调制。

从功能上看，该模块承担三重作用。首先，它为系统提供“当前是否正在接近或跨越界面”的直接物理证据；其次，它将原始低频触觉中的复杂变化压缩为可解释的三相信念分布；最后，它为策略网络提供可微、连续且可递归更新的状态条件，使控制输出能够随着介质变化而平滑过渡。因此，4.1 模块并不是一个单纯的介质分类头，而是整个跨介质策略生成链路中的状态推断核心。更准确地说，本文的介质信念推断应被理解为一个基于低频触觉证据及其时序上下文的递归状态估计过程。

**4.2** 控制导向多属性推断与不确定性感知 (Control-Oriented Multi-Attribute Inference)

除了判断“当前处于哪种介质”之外，系统还需要回答另一个对控制同样关键的问题：当前抓取对象在施力上具有什么约束特征。在跨水-气界面过程中，不同对象对力调节的响应并不相同。脆性玻璃器皿允许的安全力窗口更窄，柔顺材料对接触扰动更敏感，收口容器在出水时可能伴随更复杂的附水与姿态变化，磨砂表面则会带来不同于光滑表面的摩擦特性。因此，若策略层仅依赖对象 ID 或粗粒度类别标签，往往既难解释，也不利于泛化。基于此，本文采用\*\*控制导向多属性推断（control-oriented multi-attribute inference）\*\*来构造对象条件特征。

（1）属性轴定义

本文将对象表示为三个与抓取控制直接相关的属性轴：

Fragility / Compliance：fragile / robust / compliant

Geometry：cylindrical / bowl-like / cup-like / constricted-opening

Surface：smooth / matte

其中，Fragility / Compliance 用于描述对象对施力的容忍范围以及形变响应；Geometry 用于描述其主导几何形态及开口约束方式；Surface 用于描述表面接触条件与潜在摩擦行为。与将对象压缩为单一离散类不同，这种表示方式允许同一对象同时具有多个控制属性。例如，一个迷你玻璃花瓶可表示为“fragile + constricted-opening + smooth”，一个白色带把杯可表示为“fragile + cup-like + smooth”，一个软硅胶圆柱则可表示为“compliant + cylindrical + smooth”。这种属性组合更贴近策略真正需要感知的控制条件。

需要指出的是，上述属性均在对象实例层面进行标注，并继承到该对象的全部抓取序列与时间窗口。也就是说，它们不是窗口级动态标签，而是描述对象本体特征的稳定属性。这样做有两点好处：其一，可以避免窗口级人工属性标注带来的主观漂移；其二，可以明确区分“对象属性”与“介质状态”这两类本质上不同的变量，使后续策略建模更具因果可解释性。

（2）多头属性预测

（5）与介质信念的功能分工

前者是对象级先验，后者是状态级条件。二者共同决定最终的力调节策略。例如，两个对象都处于 Interface 阶段时，若一个是 fragile + constricted-opening，另一个是 compliant + cylindrical，则二者在期望力增益和变化节律上应当不同。也正因如此，属性推断模块不是附属分类器，而是策略层条件化的重要组成部分。

（6）Stop-Gradient 约束

为了防止策略网络通过反向传播“操纵”属性分支，使其退化为仅服务于控制输出的隐式捷径变量，本文在将 g\_W输送给策略网络之前施加 stop-gradient。其目的在于保证属性头首先忠实表达对象控制属性，而不是被策略层重新塑造成一个难以解释的中间表示。这样，属性推断模块在训练中仍然受到显式属性监督的约束，策略层则只能在此基础上利用这些属性条件，而不能反向篡改其语义含义。

综上，4.2 模块实现的是从内容表征 → 多属性控制语义 → 不确定性感知条件向量的映射。它将对象“是什么样的”这一问题转化为策略层可直接使用的控制属性表示，为后续跨介质条件化施力提供了可解释、可泛化的对象先验基础。

**5. 介质条件化策略网络 (Context-Conditioned Policy Learning)**

**本模块负责将解耦的感知表征映射为具体的机械臂抓取指令。为了解决跨域过程中物理动力学的剧烈变化（如浮力消失导致的重力突增），我们采用了一种\*\*条件化生成（Conditioned Generation）\*\*架构，使单一策略网络能够平滑地在不同介质模式间切换。**

**5.1 任务上下文聚合 (Task Context Aggregation)**

在完成语义流与任务流的表征提取后，系统需要进一步构造一个能够直接服务于策略决策的统一任务状态表示。该表示不仅应包含当前窗口内的视觉上下文与触觉接触语义，还应显式引入对象控制属性先验，从而使策略网络能够区分“当前看到了什么、触到了什么、抓的又是什么样的对象”。基于这一需求，本文将窗口级任务上下文定义为：



Ftask​(W)=Concat(hWv​, zcontent​, stop\_grad(gW​))

其中h\_W^v为视觉流输出的窗口级视觉表征，z\_{content}为语义流输出的跨介质内容表征，g\_W为 4.2 节得到的对象控制属性特征。

该设计体现了三类信息的明确分工。首先，视觉表征 z\_W^v提供宏观外观与姿态上下文，例如目标器皿的整体朝向、开口形态、局部遮挡情况以及在窗口内可见的位姿变化趋势。对于水槽器皿/容器抓取任务而言，视觉流尤其有助于感知开口器皿的轮廓结构、收口容器的外部形态、以及物体出水前后的几何外观变化。虽然视觉在浑浊水和界面扰动中可能退化，但在可见条件下，它仍能为策略提供重要的全局先验。

其次，内容表征 z\_{content}提供的是由高频触觉主导的局部接触语义。与视觉表征侧重宏观外观不同，z\_{content}更关注接触建立、微小滑移、局部纹理变化、柔顺形变及接触关系重分配等信息。由于该表征在训练中受到跨模态对齐与跨介质一致性约束，它在不同介质下仍尽量保持与对象本体相关的内容稳定性。因此，z\_{content} 实际上承担了“当前触觉感受到的对象局部状态是什么”的角色。

第三，控制属性特征g\_W则不描述当前窗口中的瞬时接触，而是提供对象级的控制条件先验。不同于直接将对象 ID 输入策略网络，g\_W 通过 Fragility / Compliance、Geometry 和 Surface 三个属性轴对对象进行可解释的控制抽象。例如，一个 fragile + constricted-opening + smooth 的迷你玻璃花瓶，与一个 compliant + cylindrical + smooth 的软质硅胶圆柱，在出水过程中显然需要不同的持握调节方式；又如，cup-like 与 bowl-like 开口器皿在附水残留、姿态变化和负载迁移上的表现也可能不同。将这类属性信息显式注入任务上下文后，策略网络可以学习“同样的介质变化，面对不同控制属性的对象应如何区别对待”。

需要强调的是，这里的g\_W在进入任务上下文时仍经过 stop-gradient 处理。这样做的目的在于保持属性分支的语义独立性，避免策略网络为了降低控制误差而反向重塑属性表示，使其退化为缺乏解释性的隐式捷径变量。换言之，策略只能“利用”属性推断结果，而不能“篡改”属性推断模块的语义。

从整体上看，任务上下文聚合的作用在于将对象的**宏观外观、局部接触语义、控制属性先验**统一编码到一个共享特征空间中。它并不是对多模态特征的简单拼接，而是在语义层面完成了三类互补信息的汇聚：

z\_W^v 提供“看到了什么”；

z\_{content} 提供“触到了什么”；

g\_W提供“这个对象在控制上意味着什么”。

因此，F\_{task}(W)为后续策略模块提供了一个更具可解释性的决策起点。随后，系统将介质信念 p(W) 作为额外的状态条件引入策略层，对该任务上下文进行阶段性调制。也就是说，5.1 节得到的 )F\_{task}(W)主要负责刻画“对象与接触本体”，而后续的p(W) 则负责刻画“当前所处的物理介质阶段”。二者共同决定最终期望力值的生成方式。

就论文叙事而言，这一步也体现了本文与“直接从多模态输入回归控制量”方法的区别：本文并不让策略网络在未经结构约束的情况下自行学习所有隐含因果，而是通过任务上下文显式拆分对象内容、控制属性与环境状态，使控制输出建立在更清晰、更可解释的表征基础之上。对于水槽器皿/容器抓取这一类对象多样、介质变化显著、控制风险差异大的任务场景而言，这种结构化任务上下文是实现稳健泛化的关键前提之一。

**5.2 基于 FiLM 的连续调制 (FiLM-based Continuous Modulation)**

**为了将介质信念 $p(W)$ 转化为对控制逻辑的调节，我们引入 Feature-wise Linear Modulation (FiLM) 层。**

**机制：**

**策略网络的主干为多层感知机（MLP）。对于第 $l$ 层的激活特征 $h\_l$，FiLM 层根据当前的介质信念 $p(W)$ 生成缩放系数 $\gamma$ 和平移系数 $\beta$ 进行仿射变换：**


**$$h\_l' = \gamma(p(W)) \odot h\_l + \beta(p(W))$$**


**其中 $\gamma(\cdot)$ 和 $\beta(\cdot)$ 是由 $p(W)$ 驱动的轻量级投影网络。**


**物理意义：**


**连续平滑： 与离散的 if-else 开关不同，FiLM 允许策略在 $p(W)$ 从水下过渡到空气的过程中，连续地“滑动”其神经元激活分布。**


**增益自适应： 在水下，$p(W) \approx [1, 0, 0]$，网络倾向于低增益模式（依赖浮力支撑）；在空气中，$p(W) \approx [0, 0, 1]$，网络自动切换至高增益模式（对抗重力）。**

### **（1）为何采用专家引导而非硬编码规则**
在许多传统方案中，跨介质抓取常通过人工设计的触发逻辑实现，例如：检测到出水后立即将夹持力增加一个固定量，或按预定义 ramp 曲线提升施力。这类方法虽然实现简单，但存在两个明显局限。其一，它们通常假设界面切换是一个清晰而瞬时的事件，难以适应真实场景中持续约 2.5 s 的过渡过程；其二，它们无法根据对象属性差异自适应调整策略，例如脆性器皿、柔顺物体与收口容器显然不应采用相同的力提升节律。基于此，本文不将“何时增力、增多少力”写成显式脚本，而是通过**专家引导的回归学习**，让策略网络从多模态状态到期望力值之间建立连续映射。
### **（2）专家标签的来源**
与早期方案中采用人工理想 ramp 不同，本文的专家监督并非一条手工设定的线性上升曲线，而是由**稳定抓取条件下的实测法向力曲线拟合得到**。具体而言，在数据采集过程中，系统记录了对象从水下抓取、经过界面到完全出水并转移至台面中心这一完整过程中的真实受力变化。对于其中抓取稳定、无明显失稳的样本，可将实测法向力信号视为“可执行且有效”的专家示范。经过平滑与拟合后，可获得对应窗口级的专家期望力标签：

F\_{expert}(W).

这种监督信号具有两个重要优点。首先，它来源于真实稳定抓取过程，因此比人工设定的理想 ramp 更贴近实际控制需求；其次，它天然反映了对象属性、介质变化和机械臂运动共同作用下的结果，使策略学习到的不是抽象脚本，而是更具物理真实性的调节规律。换言之，本文希望策略网络学到的是“在当前对象与当前介质条件下，真实可行的力应如何变化”，而不是简单模仿一个脱离数据的理想函数。
### **（3）条件化生成期望力**
在前向计算中，策略网络接收两类输入：

1. 来自 5.1 节的任务上下文 F\_{task}(W)，其中已包含视觉外观、触觉内容与对象控制属性；

2\. 来自 4.1 节的介质信念p(W)，用于描述当前处于 Water、Interface 还是 Air 阶段的概率状态。

本文通过条件化机制（包括介质信念驱动的 FiLM 调制）使策略网络根据 p(W) 对内部特征进行阶段性调节。这样，网络输出的![ref1]不再是对任务上下文的无条件回归，而是一个同时依赖**对象属性先验**与**介质状态变化**的条件输出。

这一点在界面阶段尤为重要。由于界面过渡不是瞬时事件，而是持续若干窗口的连续过程，使用离散 if-else 规则往往会造成力输出的突变；而基于 p(W) 的连续条件化，则能够使策略随着界面信念的逐步变化而平滑调整施力。与此同时，g\_W又使这种调节对不同对象具有差异化响应：面对 fragile + constricted-opening 的玻璃容器，策略应更谨慎地提升施力；面对 compliant + cylindrical 的柔顺规则体，则可允许不同的调节幅度和节律。由此，输出的 ![ref1]不只是“看到界面就加力”，而是“根据当前处于何种介质阶段、面对何种对象控制属性，生成一条合理的期望力轨迹”。
### **（4）训练目标**
策略学习采用均方误差（MSE）作为监督目标：



Lpol=∥F^des(W)−Fexpert(W)∥2.\mathcal{L}\_{pol}=\left\|\hat F^{des}(W)-F\_{expert}(W)\right\|^2.Lpol​=​F^des(W)−Fexpert​(W)​2.

其中，F\_{expert}(W)为由实测法向力曲线拟合得到的专家标签。该损失鼓励网络输出在数值上逼近真实稳定抓取过程中观测到的合理施力水平。

从训练策略上看，{L}\_{pol}不应孤立理解，而应结合前面几个模块共同考量。介质信念模块为策略提供“当前阶段”的状态依据，多属性模块为策略提供“对象约束”的控制先验，任务上下文聚合模块则提供多模态感知基础。在此基础上，{L}\_{pol}负责将这些中间表示真正拉向最终控制目标。也正因此，策略回归不是一个简单的末端头，而是本文整个结构化设计最终落脚到控制输出上的关键环节。
### **（5）学术意义与方法定位**
本文采用专家引导动力学轨迹学习，并不意味着系统在测试时仍然依赖人工脚本。相反，实测力曲线仅在训练阶段充当监督信号，用于帮助策略学习“跨介质阶段如何合理调节施力”的隐含规律；在推理阶段，系统不再访问任何显式 expert 轨迹，而是依据当前窗口感知到的多模态状态自主输出 。因此，专家信号在方法中的角色不是“运行时控制器”，而是“训练期的物理教师”。

这一点对于论文叙事十分关键：本文的目标不是用神经网络去拟合一条手工规定的增力曲线，而是通过结构化状态建模，使网络真正学会**介质变化意味着控制需求变化、对象属性差异意味着调节方式差异**这一因果关系。由此，5.3 节所学习到的并非一个与场景解耦的静态控制模板，而是一种从“对象属性 + 接触语义 + 介质信念”到“连续期望力输出”的条件化动力学映射。

**6. 训练范式与数据策略 (Training Paradigm & Data Strategy)**

**6.1 复合目标函数 (Composite Objective Function)**

本文的训练目标由多项子损失联合构成，用于分别约束跨模态对齐、跨介质表征一致性、介质状态推断、对象属性推断以及期望力回归。整体复合目标可写为：

从功能上看，{L}\_{clip}与 {L}\_{inv}主要作用于内容表征层，用于建立视觉—触觉语义对齐并提升跨介质一致性；{L}\_{med}作用于介质信念模块，用于监督系统区分 Water、Interface 与 Air 三种物理状态；{L}\_{attr}作用于控制导向多属性推断模块，用于显式约束对象属性表示的语义正确性；{L}\_{pol}则作为最终控制监督，约束策略网络输出合理的期望力值。五者共同构成从感知到控制的端到端学习目标。

在本文中，属性分支承担三个相互独立但控制上互补的属性轴预测任务。因此，属性损失 {L}\_{attr}不再写成单个交叉熵，而应定义为多头交叉熵的加权和：

其中：

{L}\_{frag}用于监督 Fragility / Compliance 属性轴，对应标签集合
{fragile, robust, compliant}；

这样设计的原因在于：对象的控制约束并非单一类别即可充分表达，而是由多个属性维度共同构成。若采用单一粗粒度 标签，则容易将“脆性约束”“主导几何”“表面接触条件”等不同控制语义压缩到同一个离散类别中，既不利于解释，也会削弱策略对未见对象的泛化能力。相反，将属性损失拆分为三条显式监督轴后，模型能够分别学习“对象是否易碎/柔顺”“对象属于哪类主导几何”“对象表面更接近光滑还是磨砂”，再通过 4.2 节定义的投影方式将其组合为统一的控制属性特征 g\_W。

进一步地，这种多头属性监督还有助于提高策略层的可解释性。对于任意一个窗口 W，若模型预测其对象为 fragile + constricted-opening + smooth，则该结果不仅可直接用于策略条件化，也可在实验分析中与人类先验一致地解释控制输出。相比直接学习一个黑箱式对象 embedding，多头属性损失使中间表示更容易被审稿人与实验设计所验证。

需要说明的是，{L}\_{attr}的监督标签来自对象实例级标注，而非窗口级人工属性标注。也就是说，同一对象实例在其全部抓取序列与窗口中共享同一组属性标签。这一设计保证了属性分支学习到的是对象本体的稳定控制特征，而不会被瞬时窗口内容扰动误导为“动态属性”。

综上，本文中的属性损失应被理解为：**将对象控制先验显式编码进学习目标，使策略层面对不同对象的响应建立在可解释的多属性基础之上。** 因而，{L}\_{attr}不是附属分类约束，而是整个条件化控制框架中的关键监督项之一。

**6.2** 掩码与三相标签细则 **(Stability-Aware Sampling Strategy)**

为了使不同训练目标使用与其语义一致的数据窗口，本文在窗口级样本构建中引入了两套互补的标注机制：

1. **主三相标签（phase label）**，用于为每个窗口赋予 Water / Interface / Air 中的一种介质阶段标签；
1. **稳定段掩码（stable mask）**，用于从主标签中进一步筛选出足够干净、远离界面边界扰动的稳定窗口，仅供部分表征学习损失使用。

### **（6）方法意义**
从更高层面看，6.2 节的双层标签设计体现了本文对“界面段既重要又危险”的基本认识。界面窗口对于最终控制至关重要，因为跨介质策略的学习正是在这一阶段发生；但界面窗口又不适合作为稳定内容表征的监督源，因为其中混杂了快速变化的介质扰动、姿态切换与接触重分配。通过主三相标签与稳定段掩码的分离，本文在数据层面显式区分了“需要保留以学习控制变化的窗口”和“需要筛除以学习稳定表征的窗口”，从而使后续各子模块的训练目标更加清晰、互不干扰。

综上，本文并非简单地将整条抓取序列粗略分为三段，而是基于样本级关键时间点、固定时间窗口与双层标签规则，构造出既适合状态识别、又适合内容对齐与策略学习的窗口级训练样本体系。这一设计为后续实验、消融分析与代码实现提供了统一而可复现的标注基础。

**6.3 标签层级体系 (Hierarchical Labeling System)**

为了兼顾内容表征的细粒度、控制策略的泛化性以及数据组织的可解释性，本文采用一种由**对象级属性、样本级条件与窗口级相位**共同构成的层级标注体系。与仅使用单一对象类别标签的做法不同，该体系将“抓的是谁”“这次试验在什么条件下进行”“当前窗口处于哪个物理阶段”三类信息明确区分开来，使不同模块能够使用与其目标语义一致的监督信号。
### **（1）对象级标签：实例身份与控制属性**
最上层为**对象级标签（object-level labels）**。
每一个物体实例被分配一个唯一的对象编号 object\_id，用于区分不同的具体对象，例如不同杯子、碗、花瓶、塑料瓶或规则体实例。对象级标签承担两个功能：

第一，object\_id 用于支持内容表征的细粒度区分与跨介质一致性学习。对于同一对象在水下稳定段与空气稳定段中出现的窗口，系统可基于 object\_id 构造正样本对，从而约束内容编码器学习到“同一对象在不同介质下仍应保持一致”的表征。

第二，对象级标签还包含三条稳定的控制属性轴，即 4.2 节中定义的：

Fragility / Compliance：fragile / robust / compliant

Geometry：cylindrical / bowl-like / cup-like / constricted-opening

Surface：smooth / matte

这些属性由人工在对象实例层面标注，并保存在对象属性表中。与 object\_id 不同，它们不用于标识“具体是哪一个物体”，而是用于表达“这个物体在控制上意味着什么”。因此，在训练中，object\_id 与控制属性虽然同属对象级信息，但服务于不同目的：前者支撑表征对齐与区分，后者支撑策略条件化与可解释控制。
### **（2）样本级标签：实验条件与结果**
第二层为**样本级标签（sample-level labels）**。
一条样本被定义为一次完整抓取试验，对应于一个视频文件、一个触觉文本文件以及该次实验的条件描述。每条样本拥有一个全局唯一的 sample\_id，采用全局递增方式编号，例如 S0001, S0002, ...。这种设计使得数据索引、窗口生成、日志记录与错误样本回查都可以基于统一主键完成，而不需要在文件名中重复编码对象类别和实验条件。

样本级标签包含如下典型字段：

sample\_id

video\_path

tactile\_path

object\_id

water\_condition

lift\_speed

placement\_variant

trial\_id

trial\_result

其中，water\_condition 用于区分 clear / turbid 等不同水体条件；lift\_speed 用于区分正常出水速度与快速出水速度；placement\_variant 与 用于记录同一不对称对象在不同摆放姿态下的试验变体；trial\_id 用于标识相同对象与相同条件组合下的重复抓取次数；trial\_result 则记录抓取结局，如 stable、partial\_slip 或 fail。
这些变量并非对象本体属性，而是单次试验才会变化的条件元数据，因此应与对象级属性明确分离。

这种“对象级属性”与“样本级条件”的分离具有重要意义。前者用于描述抓取对象本身的控制约束，后者则用于描述该次试验所处的外部执行条件。由此，模型在分析错误样本、做消融实验或统计泛化性能时，可以清晰回答性能变化究竟来自对象差异，还是来自水体环境、速度条件与摆放姿态变化。
### **（3）时间事件标签：关键时刻与窗口相位**
第三层为**时间事件标签（temporal event annotations）**。
与对象级和样本级标签不同，这一层描述的是单次试验内部的时间结构。对每条样本，至少记录以下关键时间点：

tstartt\_{start}tstart​：手指开始闭合的时刻；

tif\_entert\_{if\\_enter}tif\_enter​：界面段起点，即视觉扰动首次明显出现、物体开始预备出水的时刻；

tif\_exitt\_{if\\_exit}tif\_exit​：界面段终点，即物体完全出水的时刻；

tendt\_{end}tend​：有效结束时刻。

此外，若标注条件允许，还可补充：

### **（4）三层标签体系的分工关系**
从整体方法结构上看，这三层标签在系统中的使用分工如下：

**对象级** object\_id
支持内容表征的实例区分与跨介质一致性学习；

**对象级控制属性（fragility / geometry / surface）**
支持多属性预测与策略条件化；

**样本级条件变量（water\_condition / lift\_speed / placement / result）**
用于数据组织、实验分析与条件统计；

**时间事件标签与窗口级三相标签**
用于窗口构建、介质推断监督与稳定段筛选。

由此，本文的标签体系不再是“一个对象类别标签解决所有问题”，而是通过对象、样本、时间三个层面共同组织监督信息。
这种层级化标注不仅更适合当前的跨介质抓取任务，也与本文的模块化方法设计严格一致：对象属性描述“对象是什么样的”，时间相位描述“当前处于什么物理阶段”，样本条件描述“这次试验是在什么环境和执行条件下发生的”。

综上，6.3 节的标签层级体系为整篇方法提供了数据语义上的组织基础，使不同训练目标都能够使用与其任务匹配的监督信号，从而避免粗糙标签对表征学习与策略学习造成的混淆。

**6.4 推理与部署 (Inference & Deployment)**

在实际部署阶段，系统以滑动时间窗口的方式持续处理来自视觉与触觉的多模态输入，并在每个时刻输出当前窗口对应的期望力值。与训练阶段相同，本文仍以“夹爪开始闭合”作为统一时间原点，并基于固定窗口长度与步长在共享时间轴上组织多模态观测。该设计的核心思想是：系统不依赖单帧级同步或硬阈值逻辑，而是在窗口级状态估计的基础上，连续更新介质信念、对象控制属性和期望力输出。
### **（1）在线输入与时间组织**
部署时，视觉流以视频形式持续输入，帧率为 30 Hz；触觉流以固定周期文本流或等效的在线采样缓存形式持续输入，其实际采样周期约为 39 ms。系统维护一个长度为L=1.0s 的时间窗口缓存，并以 S=0.25s 的步长向前滑动。对于当前窗口Wt=[t, t+1.0),

系统分别收集该时间区间内全部视频帧与触觉采样点，构成窗口级视觉序列 V\_{W\_t}与触觉序列 T\_{W\_t}。

由于视觉与触觉保持各自原始采样率，窗口内部的帧数与触觉点数不必严格一致。系统不对二者做逐帧强制配准，而是分别在各自模态内部进行编码，再在窗口表征层完成融合。这样既保留了视觉与触觉各自的时间分辨率，也降低了异步采样和小尺度时延误差对推理稳定性的影响。
### **（2）在线预处理与双路触觉分解**
在每个窗口内，触觉信号首先经过与训练阶段一致的预处理流程，包括必要的去噪、归一化与基于 EMA/低通的高低频分解。由此得到：

高频触觉分量T\_{high}，用于语义流中的内容表征提取；

低频触觉分量T\_{low}，用于物理流中的介质信念推断。

这一分解保证了部署阶段的功能划分与训练阶段完全一致：高频触觉主要承载对象局部接触语义，低频触觉主要承载介质状态变化证据。这样做能够在在线推理中维持较好的可解释性与模块职责清晰性。
### **（3）前向推理流程**
对每个窗口W\_t，系统依次执行如下前向步骤：
#### **1）语义流：生成内容表征**
视觉编码器对窗口内视频序列进行时序编码，得到窗口级视觉特征 z\_{W\_t}^{v}；
内容编码器对T\_{high}进行编码，得到跨介质稳定的触觉内容表征 z\_{content}。

这里需要强调的是z\_{content} 在部署时依然**独立生成**，不接收介质信念的反馈注入。该设计继承了训练阶段对因果结构的约束：对象内容表征应尽量描述“对象本体接触语义”，而不是被当前介质状态显式污染。
#### **2）物理流：更新介质信念**
证据编码器对T\_{low} 编码得到z\_{med}，随后递归介质信念模块结合上一时刻隐藏状态，输出当前窗口的介质信念分布：



p(Wt)=[pwater(Wt), pinterface(Wt), pair(Wt)].p(W\_t)=[p\_{water}(W\_t),\ p\_{interface}(W\_t),\ p\_{air}(W\_t)].p(Wt​)=[pwater​(Wt​), pinterface​(Wt​), pair​(Wt​)].

由于采用递归时序建模，系统能够在界面附近连续累积证据，而不是被单个瞬时扰动触发不稳定跳变。这使得在线部署时的介质状态估计更加平滑，也更适合为策略层提供连续条件。
#### **3）任务流：推断对象控制属性**
多属性预测模块利用 zcontentz\_{content}zcontent​（及可选视觉上下文）输出三条属性轴上的预测分布，并构造对象控制属性特征：

在在线部署中，g\_{W\_t}表示系统对当前抓取对象控制属性的最新判断，例如其是否更接近 fragile、其主导几何是 bowl-like、cup-like、cylindrical 还是 constricted-opening，以及表面更接近 smooth 还是 matte。

### **（4）底层执行与安全约束**
策略网络输出的![ref2]并不直接等同于执行器底层命令，而是作为上层目标施力输入到底层控制器中。具体而言，系统可采用力控或阻抗控制器对该目标值进行跟踪执行，使高层策略输出最终转化为真实的夹持或末端接触调节行为。

为了提升在线部署的安全性，本文保留底层**Force Limiter** 作为保护约束。当感知不确定性过高、属性估计不可靠、或策略输出异常偏大时，力限制器将对施力范围进行裁剪，避免对脆性器皿或不稳定接触对象施加过载。这样，学习得到的策略负责提供自适应的目标力变化趋势，而底层安全模块负责保证系统不会在异常情况下突破物理安全边界。
### **（5）部署阶段的对象与状态分工**
从系统结构上看，在线推理的核心是两个条件变量的协同作用：

**对象条件**g\_{W\_t}：描述当前对象在控制上意味着什么；

**状态条件**p(W\_t)：描述当前系统正处于哪一介质阶段。

前者是相对稳定的对象先验，后者是随时间快速变化的环境状态。部署时，策略网络并不依赖对象 ID 的显式记忆，而是依据g\_{W\_t}与 p(W\_t)的组合来输出力值。由此，同样处于 Interface 阶段的两个对象，若其属性分别为 fragile + constricted-opening + smooth 与 compliant + cylindrical + smooth，系统就可以产生不同的期望力调节幅度与节律。
### **（6）方法意义**
6\.4 节所描述的部署方式，与本文整体方法设计保持了一致的结构化逻辑：

语义流负责描述对象本体接触内容；

物理流负责估计当前介质状态；

任务流负责提供对象控制属性先验；

策略层在此基础上生成连续期望力值。

因此，在线推理并非简单复现一条训练期拟合的固定轨迹，而是在窗口级多模态状态估计基础上持续生成与当前对象、当前介质阶段相匹配的控制输出。这种部署方式更适合水槽器皿/容器抓取场景中普遍存在的界面持续时间波动、对象属性多样性以及视觉退化等问题，也为后续开展真实机器人实验提供了统一、可解释的实现路径。

**6.5 数据集组织与对象池设计 (Dataset Organization and Object Pool Design)**

为了使实验设计与论文问题定义保持一致，本文对数据集不再采用“多种物体混合采集”的平铺式组织方式，而是构建一个具有明确任务语义与机制分析分工的两层数据体系：

1. **主任务对象池（Main Task Object Set）**：面向真实水槽器皿/容器抓取场景；
1. **辅助机制对象池（Controlled Object Set）**：面向表征解耦、介质推断与策略条件化的机制验证。

这种组织方式的目的在于，将“真实任务意义”与“机制可解释性”区分开来。前者用于说明本文方法为什么重要、在什么场景下有用；后者用于说明本文提出的结构性设计为什么成立、各模块的功能边界是否清晰有效。
### **（1）主任务对象池：水槽器皿/容器抓取**
本文的主实验场景被定义为**水槽器皿/容器抓取（sink-side vessel grasping）**。与早期“宽泛物体抓取”的设定不同，当前版本将研究对象收缩为日常水槽环境中常见、且确实会经历“水下抓取—界面出水—空气中短时转移”这一过程的器皿与容器类物体。该对象池主要包括：

杯类器皿

碗类器皿

花瓶类小型容器

塑料瓶、奶瓶、矿泉水瓶等瓶状容器

材质覆盖玻璃、陶瓷、塑料等典型家庭场景中常见的容器材料。之所以将主场景定义为“器皿/容器抓取”而非“餐具抓取”，是因为当前实验对象中不仅包含杯和碗，也包含花瓶、牛乳瓶、饮料瓶等不宜归入传统餐具范畴的对象。采用“器皿/容器”这一表述，既更贴近真实对象池，也更自然地强调了**开口、收口、腔体、残水与出水负载变化**这些对跨介质抓取控制真正关键的结构因素。

在该主任务对象池中，对象差异并非以日常名词类别来组织，而是通过控制导向属性体系来覆盖跨介质抓取中最关键的控制维度，包括：

**Fragility / Compliance**：fragile / robust / compliant

**Geometry**：cylindrical / bowl-like / cup-like / constricted-opening

**Surface**：smooth / matte

这种组织方式具有两个优点。首先，它能够直接回应“为什么选择这些物体”这一审稿层面常见问题：本文并非为了展示对象种类多样而随意混入杂物，而是围绕跨介质抓取中最重要的控制差异有目的地构造对象池。其次，它使对象选择与后续的策略条件化设计在语义上完全一致，即：对象池本身就是围绕策略需要感知的控制属性来组织的。

以当前实验对象为例，玻璃或陶瓷杯、碗以及迷你花瓶主要构成 fragile 组；塑料牛奶瓶和矿泉水瓶则为 robust 容器类代表；硅胶规则体则提供 compliant 类型样本。在几何层面，碗状器皿主要对应 bowl-like，杯类对应 cup-like，规则体与近柱状瓶体对应 cylindrical，而迷你花瓶与部分“口小肚大”的容器则更适合归入 constricted-opening。相比早期使用 narrow-neck 这一标签，constricted-opening 更准确地反映了这类对象“开口受限、腹部尺度大于口部尺度”的真实结构特征，也更贴近抓取控制中真正重要的几何约束。
### **（2）辅助机制对象池：规则物体机制验证集**
除了主任务对象池之外，本文还保留一组规则物体作为辅助机制对象池，用于开展更可控、更容易解释的分析实验。该对象池主要包括：

玻璃圆柱

开口圆柱

磨砂圆柱

软/硬硅胶圆柱等规则体

这类对象并不承担“真实任务场景代表”的角色，而主要用于验证本文方法中若干关键机制是否成立，例如：

高频/低频触觉分解是否真的实现了内容语义与介质证据的职责分离；

同一对象在水下稳定段与空气稳定段之间，内容表征是否保持了跨介质一致性；

多属性条件向量g\_W与介质信念 p(W)是否分别承担对象约束与阶段调制的不同角色；

在视觉退化或浑浊条件下，介质推断与属性推断的鲁棒性是否仍然成立。

保留规则体机制集的原因在于：真实器皿/容器场景虽然具有更强任务意义，但其几何、材质和接触变化往往同时发生，较难清晰地隔离单一因素的作用。而规则物体则有利于在受控条件下观察模型中间变量的行为，从而为论文中的消融实验和可解释性分析提供更干净的实验平台。

其中，object\_id 用于唯一标识对象实例，三条属性轴用于后续监督多属性预测模块，并最终构成对象控制属性特征 g\_W。
#### **样本级事件表**
样本级抓取试验信息以 sample\_events.csv 的形式维护，每一行对应一次完整抓取样本，至少包含：

sample\_id

video\_path

tactile\_path

object\_id

water\_condition

lift\_speed

placement\_variant

trial\_id

trial\_result

t\_start

t\_if\_enter

t\_if\_exit

t\_end

可选的 t\_contact\_all、t\_grasp\_stable

notes

这里，sample\_id 采用全局递增编号（例如 S0001, S0002, …），用于唯一标识一条完整抓取样本；object\_id 用于将该样本映射到对应对象实例；water\_condition、lift\_speed、placement\_variant 等字段则用于记录这次试验特有的条件变量。由此，系统能够清晰区分“对象本身是什么”和“这次试验是在什么条件下进行”的两类信息。

在程序层面，最终通过对象级表与样本级表合并生成 samples.csv，供窗口切分、Dataset 构造和模型训练统一使用。
### **（4）原始文件组织与样本关联**
在原始数据存储上，本文建议采用基于 sample\_id 的统一命名与平铺式存储方式。例如：

data/raw/videos/S0001.mp4

data/raw/tactile/S0001.txt

其中，同一 sample\_id 的视频与触觉文件同名，仅后缀不同。这样做的优点在于：

1. 可保证视频与触觉之间的一一对应关系简单明确；
1. 路径信息可直接存入样本级事件表；
1. 对象属性、实验条件和关键时间点统一由表格维护，而不被冗余编码进文件名中。

在后续代码实现中，程序只需读取 sample\_events.csv 中的相对路径字段，即可定位该样本对应的视频与触觉数据。
### **（5）采集条件与样本覆盖**
当前数据采集围绕单次完整抓取过程展开，即从夹爪开始闭合，到对象出水并转移至台面中心或有效结束点为止。样本采集过程中，本文有意识地引入若干条件变化，以提升数据对真实场景波动的覆盖能力，主要包括：

清澈水 / 浑浊水

正常出水速度 / 快速出水速度

部分俯视不对称物体的不同摆放角度

轻微扰动下的不同接触顺序与局部姿态差异

其中，大部分样本属于稳定抓取样本，用于支撑主要训练与性能评估；同时保留少量部分滑移或失败样本，用于后续风险分析与边界情况检查。需要指出的是，本文的主要策略监督仍基于稳定抓取样本中的实测法向力曲线拟合而来，因此稳定样本是当前训练体系中的主干数据来源。
### **（6）该数据组织方式的意义**
从论文结构上看，6.5 节的数据集组织方式与本文方法设计形成了严格闭环。对象池并非围绕“看起来种类很多”来扩展，而是围绕策略真正需要感知的控制差异来组织；辅助规则体并非与主任务混杂，而是专门承担机制验证角色；对象级属性表与样本级事件表则分别支持多属性推断与窗口级三相标签生成。由此，数据设计、标签体系与模型结构不再是彼此独立的三个部分，而是围绕同一个研究目标——**面向水槽器皿/容器抓取的跨介质条件化施力策略学习**——被统一组织起来。

换言之，本文的数据集组织不再是“采集了一批混合物体并尝试在其上训练模型”，而是构建了一个与方法论严格对应的实验平台：主任务对象池负责回答“方法在真实器皿/容器任务中是否有效”，辅助机制集负责回答“所提出的结构性设计是否真的学习到了预期的因果分工”。这一点对于后续实验章节的消融、可解释性分析和论文整体说服力都具有重要意义。

## **6.6 标注规范**
为保证数据集构建、窗口切分、三相标签生成以及后续训练流程的一致性，本文采用统一的分层标注规范。该规范将标注对象分为**对象级属性标注**与**样本级事件标注**两类：前者描述对象本体的稳定控制属性，后者描述一次具体抓取试验中的关键时间事件与实验条件。两类标注共同构成本文训练与评估所依赖的基础元数据。
### **（1）对象级属性标注**
对象级标注面向单个对象实例进行定义，其目标是为控制导向多属性推断模块提供稳定监督。本文采用三个属性轴来描述对象的控制相关特征：

**Fragility / Compliance**：fragile / robust / compliant

**Geometry**：cylindrical / bowl-like / cup-like / constricted-opening

**Surface**：smooth / matte

其中，Fragility / Compliance 用于描述对象对施力的容忍范围和形变响应；Geometry 用于描述对象主导几何形态及开口约束方式；Surface 用于描述表面接触条件与潜在摩擦特征。
这些属性均在**对象实例层面**标注，并在该对象的全部抓取样本中共享，不随具体窗口变化。这样做的原因在于：属性反映的是对象本体的控制特征，而不是窗口级短时状态，因此应与介质状态、抓取阶段、触觉扰动等动态变量明确区分开来。

为提高标注一致性，本文在 Geometry 轴上采用如下判定原则：

**cylindrical**：主体近似柱状，无显著收口约束

**bowl-like**：开口较大、整体较浅、呈碗状张开

**cup-like**：整体呈杯状，开口相对直接

**constricted-opening**：腹部尺度明显大于开口尺度，存在显著收口特征

相比早期使用 narrow-neck 一词，constricted-opening 更准确地覆盖了当前实验对象中“口小肚大但无明显细长颈部”的迷你花瓶及部分瓶状容器，因此被确定为最终标签定义。
### **（2）样本级事件标注**
样本级标注面向一次具体抓取试验进行定义。每条样本均对应一个视频文件与一个触觉文件，并围绕该次试验记录关键时间点和实验条件。本文统一以**夹爪开始闭合**作为共享时间原点 (t=0)，在此基础上至少为每条样本标注以下四个关键时间点：

(t\_{start})：夹爪开始闭合的时刻

(t\_{if\_enter})：界面段开始时刻，即摄像头开始出水、视觉扰动首次明显出现、物体进入预备出水阶段的最早时刻

(t\_{if\_exit})：界面段结束时刻，即物体完全出水的最早时刻

(t\_{end})：样本有效结束时刻

其中，界面区间定义为：
[
I=[t\_{if\_enter},,t\_{if\_exit}].
]
该区间将作为后续窗口级三相标签生成的基础。

除上述四个必要时间点外，本文推荐在条件允许时额外标注以下辅助事件：

(t\_{contact\_all})：三指均已完成接触的时刻

(t\_{grasp\_stable})：对象已被稳定抓住、进入稳定持握状态的时刻

这两个时间点并非当前主训练流程的必需标签，但它们对于后续的抓取建立阶段分析、触觉阶段解释、专家力曲线研究以及误差诊断均具有重要参考价值。因此，本文将其视为**推荐补标项**，而非强制必标项。
### **（3）样本级条件变量标注**
除关键事件时间点外，样本级标注还需记录一次试验中可能变化的外部条件变量。这些条件并非对象固有属性，而是这次抓取试验特有的实验设定，因此应单独归入样本级表中。本文当前建议至少记录以下字段：

water\_condition：clear / turbid

lift\_speed：normal / fast

placement\_variant：物体摆放姿态版本编号

trial\_id：相同对象、相同条件下的重复编号

trial\_result：stable / partial\_slip / fail

上述条件变量将在后续数据筛选、实验对比、结果分析与鲁棒性评估中使用。例如，界面持续时间、触觉变化速率和策略调节难度都可能受到 lift\_speed 的影响；对于俯视不对称物体，placement\_variant 则能够帮助系统区分由摆放方向变化引起的视觉和接触差异。
### **（4）标注单位与记录方式**
为避免视频帧号、触觉行号与真实时间混用导致的混乱，本文建议在最终标注表中统一使用\*\*秒（second）\*\*作为时间记录单位。
若人工标注阶段更习惯以视频帧号为基准，则可先记录对应帧号，再通过
[
t=\frac{\text{frame}}{30}
]
转换为秒，并在最终表格中保留秒值。由于触觉数据时间轴将在实现阶段依据固定采样周期自动生成，因此不建议对触觉文本流单独再手工标一套事件时间，而应统一以样本级事件表作为全系统唯一时间基准。
### **（5）标注规范的作用**
综上，本文的标注规范不是面向单帧或单窗口逐条人工赋标签，而是采用“**对象级固定属性 + 样本级关键事件**”的分层设计。前者服务于多属性推断与策略条件化，后者服务于窗口切分、三相标签生成与稳定段掩码构造。
这种标注方式既降低了人工标注成本，又保证了后续代码实现和实验分析具备统一、可复现的标签来源。

## **6.7 样本级/对象级数据组织说明（新增）**
为了使数据管理、标注维护、窗口构造与模型训练在工程上保持一致，本文采用“**对象级属性表 + 样本级事件表 + 合并后的训练总表**”三层数据组织方式。该组织方式将对象本体信息、实验条件信息和时间事件信息清晰分离，避免不同层面的语义混杂在同一张表中，从而为后续的数据预处理和模型训练提供统一入口。
### **（1）对象级数据组织**
对象级数据表用于维护对象实例本身的固定属性，建议以 object\_attributes.csv 的形式保存。
该表中的每一行对应一个唯一对象实例，至少包含以下字段：

object\_id

object\_name

fragility

geometry

surface

notes

其中，object\_id 为对象实例的唯一标识符，例如 OBJ001、OBJ002 等；object\_name 用于提供人类可读的名称描述；fragility、geometry 和 surface 对应 4.2 节中定义的三条属性轴；notes 可用于补充材质说明或特殊备注，例如“玻璃”“陶瓷”“软硅胶”等。

对象级表的核心作用是：

1. 为多属性预测模块提供监督标签；
1. 为每条抓取样本提供统一引用的对象先验；
1. 避免在样本级表中重复填写对象属性，从而减少冗余与标注漂移。
### **（2）样本级数据组织**
样本级数据表用于维护一次具体抓取试验的路径、实验条件与关键事件时间点，建议以 sample\_events.csv 的形式保存。
该表中的每一行对应一个唯一抓取样本，建议至少包含以下字段：

sample\_id

video\_path

tactile\_path

object\_id

water\_condition

lift\_speed

placement\_variant

trial\_id

trial\_result

t\_start

t\_contact\_all

t\_grasp\_stable

t\_if\_enter

t\_if\_exit

t\_end

notes

其中：

sample\_id 为样本唯一编号，采用全局递增方式，例如 S0001, S0002, S0003, …

video\_path 和 tactile\_path 分别指向该样本对应的视频文件与触觉文件

object\_id 用于将该样本与对象级表关联

water\_condition、lift\_speed、placement\_variant等字段用于描述该次试验特有的实验条件

trial\_result 表征抓取结果类型

若条件允许，可补充 t\_contact\_all 与 t\_grasp\_stable 两个辅助事件时间点

这种设计使得一条样本能够被明确理解为：

一次具体抓取试验 = 1 个视频文件 + 1 个触觉文件 + 1 个对象引用 + 1 组试验条件 + 1 组关键时间事件。
### **（3）样本编号与对象编号的职责分离**
在当前数据组织中，object\_id 与 sample\_id 承担不同角色，二者不能混用：

object\_id 表示**抓取的是哪一个对象实例**

sample\_id 表示**这是第几次具体抓取试验**

例如，可以存在：

S0001 ~ S0015 均对应 OBJ001 的 15 次抓取记录

S0016 开始对应 OBJ002 的抓取记录

这种全局递增的 sample\_id 设计可保证每条试验记录在全工程中拥有唯一主键，便于后续窗口切分、错误回查、日志记录和数据划分；而对象属性则完全通过 object\_id 与对象级表关联获得。
换言之，一个对象实例对应多条样本记录，而一条样本记录只对应一个对象实例。
### **（4）总表合并与训练入口**
在程序层面，对象级属性表与样本级事件表将通过 object\_id 进行合并，最终生成一个统一供代码读取的 samples.csv。
该合并后的总表至少包含：

样本路径信息

对象属性信息

实验条件信息

关键时间事件信息

后续的窗口切分、三相标签生成、Dataset 构造与训练脚本都应以 samples.csv 为统一入口，而不再分别直接操作 object\_attributes.csv 与 sample\_events.csv。
这种“对象表—样本表—总表”的三级组织方式，使得数据标注和模型训练之间形成清晰闭环：对象表负责提供稳定属性，样本表负责提供时间事件和条件元数据，总表负责将二者拼接为可直接处理的训练对象。
### **（5）目录结构与路径组织建议**
为避免路径混乱并支持后续代码复现，本文建议采用如下数据目录组织方式：

data/

├── raw/

│   ├── videos/

│   └── tactile/

├── annotations/

│   ├── object\_attributes.csv

│   └── sample\_events.csv

├── processed/

│   ├── samples.csv

│   ├── windows.csv

│   └── ...

└── splits/

`    `├── train.txt

`    `├── val.txt

其中：

在 sample\_events.csv 中，建议采用**相对路径**而非绝对路径记录 video\_path 和 tactile\_path。
例如：

### **（7）该数据组织方式的意义**
总的来说，样本级/对象级分层组织并不仅仅是工程便利性的考虑，更与本文方法结构高度一致：

对象级信息服务于**属性建模**；

样本级信息服务于**时间组织与三相标签生成**；

合并后的总表服务于**窗口级训练样本构造**。

因此，这一组织方式为后续实现、实验和可解释性分析提供了统一且清晰的数据基础。

## **6.8 实现时基与窗口参数说明（新增）**
为了使数据预处理、窗口切分、标签生成和在线推理在实现层面保持一致，本文对视频与触觉的时间基准、窗口参数以及窗口级标签构造方式作统一规定。本节的目的在于，将前文的方法性描述进一步落实为可执行的实现标准，从而保证代码实现、实验复现与论文叙事之间不存在偏差。
### **（1）统一时间原点**
本文将**夹爪开始闭合**定义为视频与触觉两种模态共享的时间原点，记为：
[
t=0.
]
所有样本级关键事件时间，包括 (t\_{if\_enter})、(t\_{if\_exit}) 和 (t\_{end})，均相对于这一原点记录。
采用统一原点的好处在于：

1. 避免视频帧号与触觉行号之间的直接比较；
1. 使窗口切分、三相标签生成和训练输入组织全部建立在统一时间轴之上；
1. 便于后续在视频与触觉之间进行窗口级对齐，而无需逐帧级精确同步。
### **（2）视频时基**
视觉模态以视频形式采集，帧率固定为 30 Hz。设视频第 (i) 帧对应的时间戳为：
[
t\_i^{vid}=\frac{i}{30}.
]
若实现中可直接从视频文件读取更精确的时间戳，则应优先使用真实时间戳；但在当前数据条件下，基于固定帧率构造视频时间轴已足以支撑窗口级切分与标签生成。
### **（3）触觉时基**
触觉模态以文本流形式记录，原始 txt 文件中不提供显式时间戳。经测定，触觉实际采样周期约为：
[
\Delta t\_{tac}=0.039s,
]
对应实际采样频率约为：
[
f\_{tac}\approx 25.64\text{ Hz}.
]
因此，本文不再使用早期假设的 33 Hz，而统一采用 39 ms 作为触觉的实际时间基准。设第 (j) 行触觉数据的时间戳为：
[
t\_j^{tac}=j\times 0.039,
]
若索引从 1 开始，则可写为：
[
t\_j^{tac}=(j-1)\times 0.039.
]
在当前实现中，默认不额外引入固定偏移修正项 (\Delta)，即视作视觉和触觉在起点处已完成人工对齐，并依靠窗口级建模吸收半帧到一帧量级的小尺度同步误差。
### **（4）窗口参数设定**
综合界面持续时间统计、触觉采样密度以及多模态建模稳定性考虑，本文最终采用如下固定窗口参数：

[
L=1.0s,\qquad S=0.25s,
]
其中 (L) 为窗口长度，(S) 为窗口步长。

设第 (k) 个窗口为：
[
W\_k=[t\_k^s,\ t\_k^e),
\qquad t\_k^e=t\_k^s+1.0,
]
其起始时刻满足：
[
t\_k^s=k\cdot 0.25.
]
窗口中心时刻记为：
[
t\_k^c=t\_k^s+0.5.
]

之所以选择 1.0 s / 0.25 s，而非更短或更长的窗口，是基于以下考虑：

1. 单窗口需包含足够的视觉帧和触觉点，以稳定刻画局部动态；
1. 窗口不能过长，否则会过度混合 Water / Interface / Air 三相信息；
1. 窗口不能过短，否则会放大跨模态小尺度同步误差，并降低触觉低频趋势建模能力。

在当前参数下，每个窗口通常包含约 30 帧视频和 25–26 个触觉点。对于平均约 2.5 s 的界面持续过程，该步长可产生约 6–8 个界面相关窗口，从而兼顾阶段分辨率与建模稳定性。
### **（5）窗口数据提取方式**
对于任一窗口 (W\_k=[t\_k^s,t\_k^e))，视频与触觉数据的提取均按时间区间进行，而不是按固定数量硬切：

视频窗口：
[
V\_{W\_k}={I\_i\mid t\_i^{vid}\in[t\_k^s,t\_k^e)}
]

触觉窗口：
[
T\_{W\_k}={x\_j\mid t\_j^{tac}\in[t\_k^s,t\_k^e)}
]

这种做法意味着视觉帧数与触觉点数可以天然不同，系统不强制将二者重采样到同一长度，而是在各模态内部独立时序编码后，在窗口表征层完成融合。
因此，实现上真正对齐的不是“第几帧对应第几行”，而是“同一时间窗口内的多模态局部动态”。
### **（6）界面持续时间与窗口关系**
根据样本级视频观察，本文中“界面段”通常定义为：
**摄像头开始出水、视觉扰动首次明显出现、物体进入预备出水阶段**，直到**物体完全出水**之间的区间。
该区间的持续时间平均约为：
[
2\.5s
]
（不同样本存在小幅波动）。

这一定量结果直接支撑了窗口参数的选取：

若窗口过长（如 1.5–2.0 s），则一个窗口会过度吞并大半界面过程，降低相位分辨率；

若窗口过短（如 0.3–0.5 s），则会增加噪声敏感性并削弱低频趋势建模；

采用 1.0 s 窗口既足以覆盖局部动态，又能在界面区间内产生多个连续窗口，从而支持介质信念的渐进变化建模与策略平滑调制。
### **（7）三相标签与稳定段的实现依赖**
本文所有窗口级主三相标签与稳定段掩码，均建立在上述统一时基与窗口参数基础上。也就是说：

若没有统一时间原点，视频关键事件无法稳定映射到触觉时间轴；

若没有固定采样周期，触觉 txt 行号无法转化为可用于切窗的真实时间；

若窗口长度和步长频繁变化，则后续三相标签生成规则和稳定段筛选规则也会随之漂移。

因此，本节所规定的时基和窗口参数不仅是实现细节，更是整个方法得以稳定复现的基础条件。它们决定了数据如何被组织成窗口、界面如何被分解成多个过渡样本、以及后续各损失函数将接收到什么样的训练单元。
### **（8）实现意义**
从整体上看，6.8 节的作用在于为前文的方法设计提供一个明确的工程落点：

1\.1 节定义了为什么需要窗口；

6\.2 节定义了窗口如何被赋予相位标签与稳定掩码；

而本节则明确规定了窗口到底如何在代码中切出、各模态时间轴如何构造、以及为什么当前参数选择是合理的。

由此，方法描述、数据标注和工程实现三者之间形成闭环，使本文不仅在概念上自洽，也在实现层面具备直接落地的可操作性。
